```python
import requests
import matplotlib.pyplot as plt 
from pprint import pprint
import numpy as np
```

```python
url = 'http://127.0.0.1:5555/reporting-api/meta/get-report'
obj = {
    "date_start":"2023-02-01",
    "date_stop":"2023-02-25",
    "account_id":"525510428828068",
    "fields":["impressions","clicks","total_spend","frequency"],
    "series":"impressions",
    "level":"adset"
}


response = requests.post(url,json=obj)
data = response.json()
```

```python
pprint(data[0])
```

```python
series = data[0]['series']

x = np.array([s['date'] for s in series])
y = np.array([s['impressions'] for s in series])
```

```python
fig = plt.figure(figsize=(8,5))

plt.plot(x,y)
plt.xticks(rotation=70)
plt.show()
```
